## `i` Cosine Similarity
Finish implmenting the function below to calculate `Cosine Similarity` between two vectors

```python
def cosine_sim(vec_a, vec_b):

    """
    Returns the raw cosine similarity score between two vectors.

            Parameters:
                vec_a (pandas.Series): Vector A
                vec_b (pandas.Series): Vector B

            Returns:
                sim_score (float): Similarity score between vectors vec_a and vec_b 
    """
    adj_a = vec_a 
    adj_b = vec_b
    sim_score = np.dot(adj_a, adj_b) / (np.linalg.norm(adj_a)*np.linalg.norm(adj_b))
    

    return sim_score
```

## `ii` Adjusted Cosine Similarity
Finish implmenting the function below to calculate `Adjusted Cosine Similarity` between two vectors

```python
def adjusted_cosine_sim(vec_a, vec_b):
    """
    Returns the adjusted cosine similarity score between two vectors.

            Parameters:
                vec_a (pandas.Series): Vector A
                vec_b (pandas.Series): Vector B

            Returns:
                sim_score (float): Similarity score between vectors vec_a and vec_b
    """

    adj_a = vec_a - np.mean(vec_a)
    adj_b = vec_b - np.mean(vec_b)
    sim_score = np.dot(adj_a, adj_b) / (np.linalg.norm(adj_a)*np.linalg.norm(adj_b))
    # sim_score = cosine_sim(adj_a, adj_b)

    return sim_score
```

## `iii` Pearson Correlation Coefficient
Finish implmenting the function below to calculate `Pearson Correlation Coefficient` between two vectors

```python
def pearson_sim(vec_a, vec_b):
    """
    Returns the pearson similarity score between two vectors.

            Parameters:
                vec_a (pandas.Series): Vector A
                vec_b (pandas.Series): Vector B

            Returns:
                sim_score (float): Similarity score between vectors vec_a and vec_b 
    """
    
    adj_a = vec_a - np.mean(vec_a)
    adj_b = vec_b - np.mean(vec_b)
    sim_score = np.dot(adj_a, adj_b) / (np.linalg.norm(adj_a)*np.linalg.norm(adj_b))

    return sim_score
```

## `iv` Mean Squared Difference
Finish implmenting the function below to calculate `Mean Squared Difference` between two vectors

**Note**: Make sure you calculate the difference for common dimensions only (i.e. the dimensions both items/users have non-zero values in)

```python
def msd_sim(vec_a, vec_b):
    """
    Returns the mean squared difference similarity score between two vectors.
    Note: Only consider common items between the two vectors

            Parameters:
                vec_a (pandas.Series): Vector A
                vec_b (pandas.Series): Vector B

            Returns:
                sim_score (float): Similarity score between vectors vec_a and vec_b 
    """
    diff=(vec_a - vec_b)**2
    sim_score =np.mean(diff)
    sim_score=1/(1+sim_score)
    return sim_score
```

## `iii` Top-K Candidate Selection

Selet top-K (a k of your choice) similar items for each item (user 199988) rated from the filtered similarity matrix above.

```python
#filtered_sim_df.info()
```

```python
k = 17

def select_top_k_values_excluding_ones(col):
   
    filtered_values = [value for value in col if value != 1 and value >=0]
    
    
    top_k_values = sorted(filtered_values, reverse=True)[:k]
    
    top_k_values += [np.nan] * (k - len(top_k_values))
    
    return top_k_values

top_k_values = filtered_sim_df.apply(select_top_k_values_excluding_ones)
top_k_values
```

## `iv` Candidate Rating Prediction

Calculate the predicted rating for each of the candidate items.

```python
def predict_ratings_for_candidate_items(user_id, candidate_items, user_item_matrix, item_similarity_matrix):
    predictions=[]
    user_ratings = user_item_matrix[user_id]
    for item in candidate_items:
        if item in item_similarity_matrix.columns:
            similarity_scores = item_similarity_matrix[item]
            #relevant_ratings = user_ratings.dropna()
            
            if not user_ratings.empty:
                predicted_rating = np.sum(similarity_scores * user_ratings) / np.sum(similarity_scores)
                
            else:
                predicted_rating = np.nan
            top_references = similarity_scores.sort_values(ascending=False).head(2)
            ref_items = top_references.index.tolist()
            ref_similarities = top_references.tolist()
            ref_ratings = user_item_matrix.loc[ref_items,user_id].tolist()
            
            # Append prediction and reference items details to list
            predictions.append({
                'candidate': item,
                'predicted_rating': predicted_rating,
                'ref_1': ref_items[0],
                'ref_1_similarity': ref_similarities[0],
                'ref_1_rating': ref_ratings[0],
                'ref_2': ref_items[1],
                'ref_2_similarity': ref_similarities[1],
                'ref_2_rating': ref_ratings[1]
            })
    
    # Create DataFrame from predictions list
    predictions_df = pd.DataFrame(predictions)
    return predictions_df

# Example: Predict ratings for candidate items for user 'user_3'
predicted_ratings = predict_ratings_for_candidate_items(user_id, top_k_values, utility_matrix, sim_df)
predicted_ratings.head()
```

```python
user_id = 199988

# Filter out items rated by the user
items_rated_by_user = ratings[ratings['userID'] == user_id]['songID'].tolist()


item_user_matrix = pd.pivot_table(ratings, values='rating', index='songID', columns='userID', fill_value=0)
candidate_items = ratings[~ratings['songID'].isin(items_rated_by_user)]['songID'].unique()
sim_mat = pd.DataFrame(index=utility_matrix.index, columns=utility_matrix.index)
for i in utility_matrix.index:
    for j in utility_matrix.index:
        if i != j:
            sim_mat.loc[i, j] = adjusted_cosine_sim(utility_matrix.loc[i], utility_matrix.loc[j])
        else:
            sim_mat.loc[i, j] = float(1)
sim_df = pd.DataFrame(sim_mat, index=utility_matrix.index, columns=utility_matrix.index)
item_similarity_matrix=sim_df.astype(float)

```

```python
def predict_ratings_for_candidate_items(user_id, candidate_items, item_user_matrix, item_similarity_matrix):
    
    predicted_ratings = []

    user_ratings = item_user_matrix[user_id]

    for item in candidate_items:
        if item in item_similarity_matrix.index:
            similarity_scores = item_similarity_matrix.loc[item]
            relevant_ratings = user_ratings[user_ratings.index.isin(similarity_scores.index)]

            if not relevant_ratings.empty:
                predicted_rating = np.sum(similarity_scores * relevant_ratings) / np.sum(np.abs(similarity_scores))
            else:
                predicted_rating = np.nan
            
            predicted_ratings.append({
                'candidate': item,
                'predicted_rating': predicted_rating
            })

    predicted_ratings_df = pd.DataFrame(predicted_ratings)
    k = 17
    top_k_candidates = predicted_ratings_df.nlargest(k, 'predicted_rating')

    return top_k_candidates



user_id = 199988

top_k_candidates = predict_ratings_for_candidate_items(user_id, candidate_items,item_user_matrix, item_similarity_matrix)
print("Top", k, "Recommended Songs for User", user_id, ":")
print(top_k_candidates)
```
